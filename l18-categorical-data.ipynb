# Reducing Pandas memory usage #1: lossless compression
1. How to use the ‘category’ dtype?<br> 
* Just astype('category')
2. Why should you use the ‘category’?<br> 
* Save memory (except the Gotchas situation), speed up operation performance (most likely), and improve machine learning performance (no scientific proof but there are good reasons to do so).
3. When to use it?<br> 
* Almost always, at least give it a try. Even though sometime it might take more space, but in the long run, it speeds up operation performance and can seamlessly integrate into machine learning models with categorical feature support.<br>
