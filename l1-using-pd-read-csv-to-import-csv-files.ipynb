# Changing the header 
NOTE: As the default behavior of the parameter header, the first row is selected as header (header=0)
We can change the header by passing the index value or we can set it as None



```python
#changing the heder to index 1
df2=pd.read_csv("../input/nse-tickers-their-yahoo-finance-equivalent-codes/EQUITY_L.csv",header=1)
df2
```

```python
#changing the heder to index None
df3=pd.read_csv("../input/nse-tickers-their-yahoo-finance-equivalent-codes/EQUITY_L.csv",header=None)
df3
```
