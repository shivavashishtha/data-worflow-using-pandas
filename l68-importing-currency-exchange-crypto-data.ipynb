# Initialising the FX Tickers 
We have unique tickers for Foreign Exchange too. Using those tickers we are going to fetch the data with period of year till date 

```python
t1="EURUSD=X"
t2="INR=X"
t3="RUB=X"
```

```python
eurusd=yf.download(t1,period="ytd")
```

```python
inrx=yf.download(t2,period="ytd")
```

```python
rubx=yf.download(t3,period="ytd")
```

```python
eurusd
```

```python
eurusd.info()
```

```python
from matplotlib.pyplot import figure

figure(figsize=(14, 6), dpi=80)
plt.plot(inrx.Close)
plt.legend()
plt.show()
```

```python
from matplotlib.pyplot import figure

figure(figsize=(14, 6), dpi=80)
plt.plot(eurusd.Close)
plt.legend()
plt.show()
```

```python
from matplotlib.pyplot import figure

figure(figsize=(14, 6), dpi=80)
plt.plot(rubx.Close)
plt.legend()
plt.show()
```
