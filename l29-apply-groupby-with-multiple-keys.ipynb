# How Groupby object can be used with multiple keys

```python
#importing the data
df=pd.read_csv("../input/data-science-job-salaries/ds_salaries.csv")
```

```python
#checking the dataframe generated
df
```

```python
#checking teh columns
df.columns
```

```python
#checking the unique values in job title that can be used for grouping the data
df.job_title.unique()
```

```python
#checking the number of unique values in job title
df.job_title.nunique()
```

```python
# grouping the data using job title
# This will generate 50 clusters and nunique is 50
split1=df.groupby("job_title")
```

```python
#typecasting to list
split1=list(split1)
```

```python
#checking length to varify teh results
len(split1)
```

```python
#creating another groupby object via salary column
df.salary.unique()
```

```python
#checking the nunique of salary
df.salary.nunique()
```

```python
#Applying groupby with multiple keys
split2=df.groupby(["job_title","salary"])
```

```python
#typecasting to list
split2=list(split2)
```

```python
#checking the length
len(split2)
```

```python
#slicing teh lsit to get deep into the data
split2[3][:]
```
