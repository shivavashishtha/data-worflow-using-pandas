# Producing boolean series using .duplicated()

```python
df["2021 Squad"].duplicated(keep=False)
```

```python
# This is mark the first occurrence of the duplicated values as false
# This argument will treat the first Occurrence as genuine elemets by marking it false
df["2021 Squad"].duplicated(keep="first")
```

```python
# This is mark the last occurrence of the duplicated values as false
# This argument will treat the last Occurrence as genuine elemets by marking it false
df["2021 Squad"].duplicated(keep="last")
```

# Keep 
keep : {'first', 'last', False}, default 'first'
    Determines which duplicates (if any) to mark.

    - ``first`` : Mark duplicates as ``True`` except for the first occurrence.
    - ``last`` : Mark duplicates as ``True`` except for the last occurrence.
    - False : Mark all duplicates as ``True``.
