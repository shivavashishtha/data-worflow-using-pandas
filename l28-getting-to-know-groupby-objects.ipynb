# pandas.DataFrame.groupby
A groupby operation involves some combination of splitting the object, applying a function, and combining the results. This can be used to group large amounts of data and compute operations on these groups. Used to determine the groups for the groupby.


![m](https://datascienceparichay.com/wp-content/uploads/2021/10/pandas-groupby-standard-dev.png.webp)

![m](https://pbpython.com/images/groupby-example.png)

```python
#using groupby
x=df.groupby("species")
```

```python
#checking the groupby object
x
```

```python
#checking different functions applicable for the groupby object
dir(x)
```

```python
#checking the groups of groupby object
x.groups
```

```python
#typecasting the groupby object to list object
xl=list(x)
```

# Manual subsitute of Groupby Object

```python
#checking the unique species that can be used for groupby operation
df.species.unique()
```

```python
#.loc boolean array can be used as subsitute of groupby
# The limitation of .loc bollean array is that it used manually and can not be used when we have the clusters in 1000s
df.loc[df.species=="Iris-virginica"]
```

```python
df.loc[df.species=="Iris-versicolor"]
```
