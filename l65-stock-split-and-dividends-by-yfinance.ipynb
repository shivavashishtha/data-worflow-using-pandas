# Stock Split and Dividends by YFinance
These type of data can also be fetched using yfinance library by setting actions argument true<br><br>
**actions: bool**
* Download dividend + stock splits data. Default is False

```python
#installing the library
!pip install yfinance
```

```python
#importing the library in the environment
import yfinance as yf
```

```python
#setting the ticker
ticker="aapl"
```

```python
#creating a dataframe by fetching the data 
df=yf.download(ticker,period="10y",actions=True)
```

```python
#checking the dataframe
df
```

```python
#checking the info
df.info()
```

```python
df.columns
```

```python
#producing a dataframe with dividend values of more than 0
df[df.Dividends>0]
```

```python
#checking the data of specific range 
df.loc["2012-08-05":"2012-08-15"]
```

```python
df.loc["2012-08-05":"2012-08-15"].diff()
```

```python
#producing a dataframe with stock splits values of more than 0
df[df["Stock Splits"]>0]
```
