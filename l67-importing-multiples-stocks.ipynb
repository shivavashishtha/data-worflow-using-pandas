# Importing multiples stocks and Financial Indexes Data by YFinance

```python
#creating a list of multiple tickers
tickers=["SBI","TCS"]
```

```python
#installing yfinance library
!pip install yfinance
```

```python
#importing the library
import yfinance as yf
```

```python
#downloading the data with 1 year period and grouping via ticker
df=yf.download(tickers,period="1y",actions=True, group_by="Ticker")
```

# Fetching Indeces Data
we can use the same method to fetch the data of the indeces as we know the ticker of indeces will also be unique. Therefore, we can fetch the data of the indeces using the ticker and ^sign in front of the ticker

```python
#fetching nse bank data of 1year period
nsebank=yf.download("^NSEBANK",period="1y")
```

```python
nsebank.columns
```

```python
#plotting the close price
from matplotlib.pyplot import figure

figure(figsize=(14, 6), dpi=80)
plt.plot(nsebank.Close)
plt.legend()
plt.show()
```

```python
#fetching BSE 30 data of 1year period
bse30=yf.download("^BSESN",period="1y")
```

```python
#plotting the close price
from matplotlib.pyplot import figure

figure(figsize=(14, 6), dpi=80)
plt.plot(bse30.Close)
plt.legend()
plt.show()
```
